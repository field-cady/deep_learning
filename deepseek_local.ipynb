````python
import ollama
import sys

ollama.list()

#MODEL = 'deepseek-r1:1.5b' # sucks ass
MODEL = 'llama3.2:3b'
````

````python
response = ollama.chat(model=MODEL, 
                     messages = [{'role':'user',
                                  'content': 'Show me a "hello world" script in python'}
                                ]
                    )

print(response['message']['content'])
````

````python
stream = ollama.chat(model=MODEL,
                     messages = [{'role':'user',
                                  'content': 'What is "red" in Spanish?'},
                                 {'role':'user',
                                  'content': 'What is "green" in Spanish?'},
                                ],
                     stream=True
                    )
for chunk in stream:
    print(chunk['message']['content'], end='', flush=True)
````
