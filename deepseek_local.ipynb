# Experimenting w running DeepSeek locally via Ollama

Make sure to first run 
```
ollama run deepseek-r1:1.5b
```

````python
import ollama
import sys

ollama.list()
````

````python
response = ollama.chat(model='deepseek-r1:1.5b', 
                     messages = [{'role':'user',
                                  'content': 'Show me a "hello world" script in python'}
                                ]
                    )

print(response['message']['content'])
#for chunk in stream:
#    print(chunk['message']['content'], end='', flush=True)
````

````python
stream = ollama.chat(model='deepseek-r1:1.5b', 
                     messages = [{'role':'user',
                                  'content': 'What is "red" in Spanish?'},
                                 {'role':'user',
                                  'content': 'What is "green" in Spanish?'},
                                ],
                     stream=True
                    )
for chunk in stream:
    print(chunk['message']['content'], end='', flush=True)
````
